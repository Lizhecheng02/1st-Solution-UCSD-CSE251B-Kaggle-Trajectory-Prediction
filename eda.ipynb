```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_file = np.load("./data/train.npz")
train_data = train_file["data"]
print("train_data shape", train_data.shape)

test_file = np.load("./data/test_input.npz")
test_data = test_file["data"]
print("test_data shape", test_data.shape)
```

```python
def count_non_zero_agents(data):
    non_zero_counts = []
    for scene in data:
        agent_mask = np.any(scene != 0, axis=(1, 2))
        count = np.sum(agent_mask)
        non_zero_counts.append(count)
    return non_zero_counts


train_non_zero_counts = count_non_zero_agents(train_data)
test_non_zero_counts = count_non_zero_agents(test_data)
```

```python
plt.figure(figsize=(12, 6))
plt.hist(train_non_zero_counts, bins=range(0, 52), edgecolor="black")
plt.title("Non-zero Agents per Scene in Training Data")
plt.xlabel("Number of Non-zero Agents")
plt.ylabel("Number of Scenes")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(test_non_zero_counts, bins=range(0, 52), edgecolor="black")
plt.title("Non-zero Agents per Scene in Test Data")
plt.xlabel("Number of Non-zero Agents")
plt.ylabel("Number of Scenes")
plt.grid(True)
plt.show()
```
